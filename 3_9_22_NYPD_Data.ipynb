```python
# I am going to remove the ~20 outlier arrests in Manhattan police precinct 114.
# I am doing this because it ruins my ability to bin the arrests by precinct and group them by borough
df2 = df[df.ARREST_PRECINCT != 114]
df2
```

```python
# Verifying that precinct 114 was dropped from ARREST_PRECINCT in the dataframe.
condition = df2["ARREST_BORO"] == 'M'
dfm2 = df2[condition]
dfm2.sort_values(by = 'ARREST_PRECINCT').tail(50)

# The table now ends with ARREST_PRECINCT 34, which is what I wanted.
```

```python
# Police precinct ranges
# 1-34 Manhattan
# 40-52 Bronx
# 60-94 Kings
# 100-115 Queens
# 120-123 Staten Island

# Creating a dictionary to allow me to iterate across the four columns I'm interested in (age, sex, race, & offense)
arrests_by_group = {}
groups = ["AGE_GROUP", "PERP_SEX", "PERP_RACE", "OFNS_DESC"]

# Placing the range of police precincts across the 5 boroughs into 5 bins (0-39, 40-59, 60-99, 100-119, 120-129).
# This creates a new "Precinct Borough" column. By including "ARREST_KEY" I'm able to obtain a count of arrests by offense and demographic group.
# I've ordered the groupby so the largest values appear at the top.
# I will limit my results to the 5 most common offenses and allow up to 10 results for each of the demographic categories.
# I renamed "ARREST_KEY" to "Arrests"
df["PRECINCT_BOROUGH"] = pd.cut(df2["ARREST_PRECINCT"], bins = [0, 39, 59, 99, 119, 129], labels = ["Manhattan", "Bronx", "Kings", "Queens", "Staten Island"])
for group in groups:
  temp_df = df2[["PRECINCT_BOROUGH", group, "ARREST_KEY"]]
  temp_df = temp_df.groupby(by = ["PRECINCT_BOROUGH", group]).count()
  temp_df = temp_df.sort_values(by = ["PRECINCT_BOROUGH", "ARREST_KEY"], ascending = False).groupby("PRECINCT_BOROUGH").head(5 if group == "OFNS_DESC" else 10)
  temp_df = temp_df.rename(columns = {"ARREST_KEY": "Arrests"})
  arrests_by_group[group] = temp_df
```

```python
# Examining the five most common offenses for arrests across the five boroughs
arrests_by_group["OFNS_DESC"]

# Assault 3 & Related Offenses were the most common offense committed by those arrested across every borough
# Regarding the second most common offense, the boroughs diverged
# Felony Assault was the second most common offense in Queens, Kings, & Bronx
# Felony Assault was only the 4th most common offense in Manhattan & the 5th most common offense in Staten Island
# Miscellaneous Penal Law was the second most common offense in Staten Island while Petit Larceny was the second most common offense in Manhattan
# Miscellaneous Penal Law was the third most common offense in both Queens & Kings. Conversely, this didn't make the top 5 in Bronx or Manhattan.
# Dangerous Drugs was the third most common offense in Staten Island, Bronx, & Manhattan, but this offense wasn't present in the top 5 of Queens or Kings
```

```python
# The above table appears below graphiclly as a bar chart
arrests_by_group["OFNS_DESC"].plot(kind="bar")
```

```python
# Examining arrests by age group across the five boroughs
arrests_by_group["AGE_GROUP"]
# Across all boroughs, 25-44 year olds represent the age group with the most arrests. 
# 25-54 year olds were arrested at a level 3 times higher than the second highest age group in every borough.
# The boroughs diverge between 18-24 year olds and 45-64 year olds as the 2nd most commmon age group that was arrested
# In SI, K, & B, 18-24 year olds were arrested at slightly higher levels than 25-64 year olds
# Its shocking that a group with an age range of only seven years were arrested at levels that exceeded an age range of 20 years (45-64)
# Manhattan is the only borough where 45-64 year olds were solidly the second most arrested age group (9,000 compared to 6,000 18-24 year olds)
# The boroughs were consistent with younger than 18 and older than 65 ranking 4th and 5th respectively
# Arrests among the "youth" and "elderly" were significantly lower than arrests among the "working age" population (18-64 year olds)
```
